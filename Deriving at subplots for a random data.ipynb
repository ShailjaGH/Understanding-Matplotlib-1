```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
import numpy as np
x=np.linspace(0,10,25)
y=x*x+2
print (x)
print(y)
print(np.array[x,y].reshape(25,2))
```

```python
pylab.plot(x,y,'g')
```

```python
pylab.subplot(1,2,1) #the contents of brackets represents (rows,column,index)
pylab.plot(x,y,'r--')
pylab.subplot(1,2,2)
pylab.plot(y,x,'g*-')
```

```python
pylab.subplot(1,2,1)
pylab.plot(x,y,'r--')
pylab.subplot(1,2,1)
pylab.plot(y,x,'b--')
```

```python
from matplotlib import pyplot as plt
fig=plt.figure()
axis=fig.add_axes([0.5,0.1,0.8,0.8])
axis.plot(x,y,'r')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for ax in axes:
    ax.plot(x,y,'r')
```

```python
fig=plt.figure(figsize=(16,9),dpi=300)
axes1=fig.add_axes([0.1,0.1,0.8,0.8]) #left,right,width,height of canvas
axes2=fig.add_axes([0.1,0.5,0.3,0.3])
axes1.plot(x,y,'r')
axes2.plot(y,x,'y')
```

```python
fig,axes=plt.subplots()
axes.set_title("Title of Graph")
axes.set_xlabel("x-label")
axes.set_ylabel("y-label")
axes.plot(x,y,'r')
```
